```python
#W5A1
def Max_Of_Two( a, b) :
    if a > b:
        return a
    elif a==b:
        return a
    else:
        return b

a,b = map(int,input().split())
print(Max_Of_Two(a,b))
```

```python
#W5A2
def swap(a,b):
    a , b = b, a
    return a ,b

a,b = map(int,input().split())
print(*swap(a,b))
```

```python
#W5A3
import math
def songuyento(n):
    if n < 2:
        return False
    else:
        for i in range(2, int(math.sqrt(n))+1):
            if n % i == 0:
               return False
        return True

n=int(input())
print(songuyento(n))
```

```python
#w5A4
def sohoanhao(n):
    s=0
    for i in range(1,n+1):
        if n % i == 0:
            s += i
        if s == n:
            return True
    return False

print(sohoanhao(28))
```

```python
#W5A5
def timso(danhsach,a):
    for i in range(len(danhsach)):
        if danhsach[i] == a:
            return danhsach[i]
    return -1

danhsach=[1,2,3,4,-6,12,45,67,89]
a = 5
print(timso(danhsach,a))
```

```python
#W5A6
def giaithua(n):
    t= 1
    for i in range(1,n+1):
        t *=i
    return t


print(giaithua(14))
```
