```python
#Bai2
a=float(input("Chieu dai:"))
print(a)
b=float(input("Chieu rong:"))
print(b)
pi=3.14
S_chunhat=a*b
S_hinhtron=(pi*(b/2)**2)
S_conlai=S_chunhat - S_hinhtron
#Dien tich khu vuon hinh chu nhat
print("Diện tích hình chữ nhật:",round(S_chunhat,2))
#Dien tich hinh tron
print("Diện tích hinh tron",round(S_hinhtron,2))
#Dien tich phan con lai
print("Dien tich phan con lai",round(S_conlai,2))
```

```python
#Bai3
c = input("Nhập vào một ký tự: ")
print(c)
if c.isupper():
    print("Chữ thường tương ứng là:", c.lower())
else:
    print("Chữ hoa tương ứng là:", c.upper())
```
