```python
#Bai2
a=float(input("Chieu dai:"))
print(a)
b=float(input("Chieu rong:"))
print(b)
pi=3.14
S_chunhat=a*b
S_hinhtron=(pi*(b/2)**2)
S_conlai=S_chunhat - S_hinhtron
#Dien tich khu vuon hinh chu nhat
print("Diện tích hình chữ nhật:",round(S_chunhat,2))
#Dien tich hinh tron
print("Diện tích hinh tròn:",round(S_hinhtron,2))
#Dien tich phan con lai
print("Diện tích phần còn lại:",round(S_conlai,2))
```

```python
#Bai6
a=float(input("Canh thu nhat:"))
b=float(input("Canh thu hai:"))
c=float(input("Canh thu ba:"))
print(a)
print(b)
print(c)
if a+b>c and a+c>b and b+c>a:
    #Nua chu vi tam giac
    p=(a+b+c)/2
    # Tính diện tích theo công thức Heron
    S = (p*(p-a)*(p-b)*(p-c))**0.5
    print("Diện tích tam giác là:", round(S,1))
else:
    print("Khong phai 3 canh cua tam giac")
```
