```python
#Bài 4
h = int(input())
for i in range(1, h + 1):
    for j in range(h - i):
        print(" ", end="")
    if i == 1:
        print("*")
    elif i == h:
        for j in range(2 * h - 1):
            print("*", end="")
        print()
    else:
        print("*", end="")
        for j in range(2 * i - 3):
            print(" ", end="")
        print("*")
```

```python
#Bài 9
a = int(input())
i = 0
while i * i < a:
    i += 1
if i * i == a:
    print("true")
else:
    print("false")
```

```python
#Bài 16
n = int(input())
chan = 0
le = 0
while n > 0:
    d = n % 10
    if d % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10
print(chan, le)
```
