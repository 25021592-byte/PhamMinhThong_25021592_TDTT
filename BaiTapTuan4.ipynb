```python
#W4A6
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
a, b = map(int, input().split())
tong = 0
for x in range(a, b + 1):
    if is_prime(x):
        tong += x
print(tong)
```

```python
#W4A7
def ham(x):
    if x < 2:
        return False
    for i in range(2, x):
        if x % i == 0:
            return False
    return True
n = int(input("Nhập n: "))
max_uoc = -1  
for i in range(1, n + 1):
    if n % i == 0:        
        if ham(i):   
            max_uoc = i   
print("Ước số nguyên tố lớn nhất của", n, "là:", max_uoc)
```

```python
#W4A8
def reverse(x):
    return int(str(x)[::-1])
def palindrome(x):
    return str(x) == str(x)[::-1]
n = int(input("Nhập n: "))
steps = 0  
while not palindrome(n):
    n = n + reverse(n)
    steps += 1
print("Số bước thực hiện:", steps)
print("Số palindrome thu được:", n)
```
