```python
#W4A1
#dÃ¹ng for
n = int(input())
if n > 0 and n <= 1000:
    tong = 0
    for i in range(1 ,n+1):
        tong += i
    print(tong)
else:
    print("Nhap sai du lieu")
#dÃ¹ng while
n = int(input())
if n > 0 and n <=1000:
    tong = 0
    i = 1
    while i <= n:
        tong += i
        i += 1
        print (tong ,end =' ')
else:
    print("Nhap sai du lieu")
```
