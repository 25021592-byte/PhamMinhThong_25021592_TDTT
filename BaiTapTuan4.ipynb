```python
#W4P1
#dùng for
n = int(input())
tong = 0
for i in range(1 ,n+1):
    tong += i
    print(tong)
#dung while
n = int(input())
tong = 0
i = 1
while i <= n:
    tong += i
    i += 1
    print (tong)
```

```python
#W4P2
n = int(input())
if n < 2:
    print("Lỗi")
else:
    snt = True
    for i in range( 2 ,int((n**0.5)+1)):
        if n % i == 0:
            snt = False
            break
    if snt:
        print("Là số nguyên tố")
    else:
        print("Lỗi")
```
