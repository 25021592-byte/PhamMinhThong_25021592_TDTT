```python
#Bai3
a = float(input("Nhap a: "))
b = float(input("Nhap b: "))
c = float(input("Nhap c: "))
if a+b>c and a+c>b and b+c>a:
    print("Day la tam giac")
    if a==b==c:
        print("Tam giac deu")
    elif a==b or b==c or a==c:
        print("Tam giac can")
    elif a**2+b**2==c**2 or c**2+b**2==a**2 or a**2+c**2==b**2 :
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")
    p = (a + b + c)/2
    S = (p*(p-a)*(p-b)*(p-c))**0.5
    print("Chu vi:", a+b+c)
    print("Dien tich:", S)    
else:
    print("Khong la ba canh cua mot tam giac")
```

```python
#Bai5
a = float(input("Nhap a: "))
b = float(input("Nhap b: "))
c = float(input("Nhap c: "))
if a==0:
    if b==0:
        print("Vo nghiem" if c!=0 else "Vo so nghiem")
    else:
        print("Nghiem =", -c/b)
else:
    delta = b**2 - 4*a*c
    if delta<0:
        print("Vo nghiem")
    elif delta==0:
        print("Nghiem kep x =", (-b)/(2*a))
    else:
        x1 =(-b+((delta)**0.5))/(2*a)
        x2 =(-b-((delta)**0.5))/(2*a)
        print("x1 =", x1,"x2 =", x2)
```

```python
#Bai10
a = float(input("Tâm a = "))
b = float(input("Tâm b = "))
R = float(input("Bán kính R = "))
x = float(input("x = "))
y = float(input("y = "))
d2 = (x - a)**2 + (y - b)**2
R2 = R**2
if d2 < R2:
    print("A nằm trong đường tròn")
elif d2 == R2:
    print("A nằm trên đường tròn")
else:
    print("A nằm ngoài đường tròn")
```
